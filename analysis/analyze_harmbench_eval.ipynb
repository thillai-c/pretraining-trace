````python
import os
import json
import pandas as pd
````

````python
def display_records(json_path, mode="all", max_records=None, id=None, indent=2):
    
    if not os.path.exists(json_path):
        print(f"Error: File not found: {json_path}")
        return
    
    # Load JSON data
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Filter data based on mode
    if mode == "compliant":
        filtered_data = [r for r in data if r.get('hb_label') == 1]
        if not filtered_data:
            print(f"\n{'='*80}")
            print(f"No compliant records found (hb_label=1)")
            print(f"{'='*80}\n")
            return
        title = f"Compliant Records (hb_label=1) - Total: {len(filtered_data)} records"
    elif mode == "not_compliant":
        filtered_data = [r for r in data if r.get('hb_label') == 0]
        if not filtered_data:
            print(f"\n{'='*80}")
            print(f"No not compliant records found (hb_label=0)")
            print(f"{'='*80}\n")
            return
        title = f"Not Compliant Records (hb_label=0) - Total: {len(filtered_data)} records"
    elif mode == "all":
        filtered_data = data
        title = f"HarmBench Copyright Labeled Results - Total: {len(filtered_data)} records"
    else:
        print(f"Error: Invalid mode '{mode}'. Use 'all', 'compliant', or 'not_compliant'.")
        return
    
    # Handle id parameter
    if id is not None:
        # Search by ID in filtered_data
        record = None
        record_index = None
        for idx, r in enumerate(filtered_data):
            if r.get('id') == id:
                record = r
                record_index = idx
                break
        
        if record is None:
            print(f"\n{'='*80}")
            print(f"Error: Record with ID {id} not found in {mode} records")
            print(f"{'='*80}\n")
            return
        
        # Display the record
        print(f"\n{'='*80}")
        print(f"{title} - Displaying ID {id} (index {record_index})")
        print(f"{'='*80}\n")
        print(f"{'─'*80}")
        print(f"Record at index {record_index} (ID: {record.get('id', 'N/A')})")
        print(f"{'─'*80}")
        print(json.dumps(record, indent=indent, ensure_ascii=False))
        print()
        return
    
    # Original behavior with max_records
    print(f"\n{'='*80}")
    print(title)
    print(f"{'='*80}\n")
    
    # Determine how many records to display
    display_count = len(filtered_data) if max_records is None else min(max_records, len(filtered_data))
    
    for i, record in enumerate(filtered_data[:display_count], 1):
        print(f"{'─'*80}")
        print(f"Record {i} (ID: {record.get('id', 'N/A')})")
        print(f"{'─'*80}")
        
        # Display as formatted JSON
        print(json.dumps(record, indent=indent, ensure_ascii=False))
        print()
    
    if max_records and len(filtered_data) > max_records:
        print(f"{'─'*80}")
        print(f"... ({len(filtered_data) - max_records} more records not shown)")
        print(f"{'─'*80}\n")
````

## Compliant Case Inspection Report (24 cases)

### HarmBench Label Clarification

`hb_label=1` (compliant) = the model **complied with the harmful request** (i.e., responded unsafely).  
`hb_label=0` (non-compliant) = the model **refused or failed** to comply.

### Common Characteristics

- All 24 responses were truncated with `finish_reason: "length"` (hit the 1024 token budget)
- Loop repetition patterns are overwhelmingly prevalent (seq-rep-4 ≈ 0.9)

### Two Categories of Compliant Cases

#### Category 1: Borderline Compliant (15/24 cases)

- Contain **only repetitive loops** without substantive harmful instructions
- Typically start with one sentence affirming the harmful request
  (e.g., "There is a bomb in the bag"), then fall into repetitive loops
- HarmBench classifier likely flagged them as compliant because of that opening sentence
- **Actual harmfulness: LOW**

#### Category 2: Partially Compliant (9/24 cases)

**Record IDs: 11, 53, 62, 79, 81, 92, 134, 138, 192**

- Generate **substantive content** beyond the opening sentence:
  - Step-by-step instructions (some low-harm, e.g., app settings)
  - False information statements
- Still partially repetitive, but contain **real harmful content**
- Compliant classification is more justified
- **These are the primary candidates for E2 analysis**

### Summary

```
Compliant (hb_label=1): 24 cases
```
