```python
import os
import json
import pandas as pd
```

```python
# Load and display labeled JSON results
labeled_json_path = "./results/gpt_j_6b/harmbench_standard_labeled.json"

if os.path.exists(labeled_json_path):
    with open(labeled_json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

# Print summary statistics
print(f"\n{'='*80}")
print(f"Summary Statistics")
print(f"{'='*80}")

total = len(data)
compliant = sum(1 for r in data if r.get('hb_label') == 1)
not_compliant = sum(1 for r in data if r.get('hb_label') == 0)
with_errors = sum(1 for r in data if 'error' in r)
no_label = sum(1 for r in data if 'hb_label' not in r and 'error' not in r)

print(f"Total records: {total}")
print(f"Compliant (hb_label=1): {compliant} ({compliant/total*100:.1f}%)")
print(f"Not Compliant (hb_label=0): {not_compliant} ({not_compliant/total*100:.1f}%)")
if with_errors > 0:
    print(f"Records with errors: {with_errors}")
if no_label > 0:
    print(f"Records without label: {no_label}")
print(f"{'='*80}\n")
```

```python
def display_records(json_path, mode="all", max_records=None, index=None, indent=2):

    if not os.path.exists(json_path):
        print(f"Error: File not found: {json_path}")
        return
    
    # Load JSON data
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Filter data based on mode
    if mode == "compliant":
        filtered_data = [r for r in data if r.get('hb_label') == 1]
        if not filtered_data:
            print(f"\n{'='*80}")
            print(f"No compliant records found (hb_label=1)")
            print(f"{'='*80}\n")
            return
        title = f"Compliant Records (hb_label=1) - Total: {len(filtered_data)} records"
    elif mode == "not_compliant":
        filtered_data = [r for r in data if r.get('hb_label') == 0]
        if not filtered_data:
            print(f"\n{'='*80}")
            print(f"No not compliant records found (hb_label=0)")
            print(f"{'='*80}\n")
            return
        title = f"Not Compliant Records (hb_label=0) - Total: {len(filtered_data)} records"
    elif mode == "all":
        filtered_data = data
        title = f"HarmBench Copyright Labeled Results - Total: {len(filtered_data)} records"
    else:
        print(f"Error: Invalid mode '{mode}'. Use 'all', 'compliant', or 'not_compliant'.")
        return
    
    # Handle index parameter
    if index is not None:
        # Check if index is a range (tuple or list with 2 elements)
        if isinstance(index, (tuple, list)) and len(index) == 2:
            start_idx, end_idx = index
            # Validate range
            if start_idx < 0 or end_idx >= len(filtered_data) or start_idx > end_idx:
                print(f"\n{'='*80}")
                print(f"Error: Invalid range [{start_idx}, {end_idx}]. Valid range: 0 to {len(filtered_data) - 1}")
                print(f"{'='*80}\n")
                return
            
            # Display records in the specified range
            print(f"\n{'='*80}")
            print(f"{title} - Displaying indices {start_idx} to {end_idx} (inclusive)")
            print(f"{'='*80}\n")
            
            for i in range(start_idx, end_idx + 1):
                record = filtered_data[i]
                print(f"{'─'*80}")
                print(f"Record at index {i} (ID: {record.get('id', 'N/A')})")
                print(f"{'─'*80}")
                print(json.dumps(record, indent=indent, ensure_ascii=False))
                print()
            return
        
        # Single index (original behavior)
        elif isinstance(index, int):
            if index < 0 or index >= len(filtered_data):
                print(f"\n{'='*80}")
                print(f"Error: Index {index} is out of range. Valid range: 0 to {len(filtered_data) - 1}")
                print(f"{'='*80}\n")
                return
            
            # Display only the specified index
            record = filtered_data[index]
            print(f"\n{'='*80}")
            print(f"{title} - Displaying index {index}")
            print(f"{'='*80}\n")
            print(f"{'─'*80}")
            print(f"Record at index {index} (ID: {record.get('id', 'N/A')})")
            print(f"{'─'*80}")
            print(json.dumps(record, indent=indent, ensure_ascii=False))
            print()
            return
        else:
            print(f"\n{'='*80}")
            print(f"Error: Invalid index format. Use an integer (e.g., 10) or a range tuple/list (e.g., (10, 20))")
            print(f"{'='*80}\n")
            return
    
    # Original behavior with max_records
    print(f"\n{'='*80}")
    print(title)
    print(f"{'='*80}\n")
    
    # Determine how many records to display
    display_count = len(filtered_data) if max_records is None else min(max_records, len(filtered_data))
    
    for i, record in enumerate(filtered_data[:display_count], 1):
        print(f"{'─'*80}")
        print(f"Record {i} (ID: {record.get('id', 'N/A')})")
        print(f"{'─'*80}")
        
        # Display as formatted JSON
        print(json.dumps(record, indent=indent, ensure_ascii=False))
        print()
    
    if max_records and len(filtered_data) > max_records:
        print(f"{'─'*80}")
        print(f"... ({len(filtered_data) - max_records} more records not shown)")
        print(f"{'─'*80}\n")
```

```python
display_records(labeled_json_path, mode="all", max_records=3)
```

```python
display_records(labeled_json_path, mode="compliant", max_records=3)
```

## Report

In HarmBench, `hb_label=1` (compliant) means "the model actually complied with the harmful request".  
In other words, this does not indicate that safety was maintained; instead, cases where the model responded unsafely are classified as compliant.

**Common characteristics**: 
- All responses were truncated with `finish_reason: "length"`
- Loop repetition patterns are overwhelmingly prevalent

Most are not truly compliant (borderline cases). They contain only repetitive loops *without* substantive instructions.  
However, because most begin by affirming harmful content in the opening sentence (e.g., 'There is a bomb in the bag'), the HarmBench judge likely classified them as compliant. In these cases, the actual harmfulness is *low*.

However, `Indices 1, 6, 8, 11, 13, 16, 20, 21, 23` are partially compliant. Unlike pure loop repetitions, these cases generate substantive content—either step-by-step instructions (low-harm, e.g., app settings) or false information statements. The compliant classification is more justified as they produce actual harmful content, even if partially repetitive.
