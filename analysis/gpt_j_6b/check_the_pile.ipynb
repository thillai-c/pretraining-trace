```python
import json

# JSONL: 한 줄이 하나의 JSON 객체. 줄 단위로 읽어야 함 (대용량 파일에 적합)
path = '../data/train/00.jsonl'

# 방법 1: 처음 N줄만 읽어서 확인
records = []
with open(path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i >= 5:  # 처음 5개만
            break
        records.append(json.loads(line))

# 첫 번째 레코드의 키 확인
if records:
    print(records[0].keys())
```

```python
# JSONL 파일 총 줄 수 (레코드 수) 확인
# 대용량이라 메모리에 올리지 않고 한 줄씩만 읽음

path = 'train/00.jsonl'
count = 0
with open(path, 'r', encoding='utf-8') as f:
    for _ in f:
        count += 1

print(f'Total {count:,} lines')
```
